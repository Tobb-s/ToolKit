```
# ==========================================
# PDF SPLITTER - HERRAMIENTA DE DESGUACE
# ==========================================
# Instrucciones: Copia este código en una celda
# de Google Colab y ejecutalo (Play).

import os
try:
    from PyPDF2 import PdfReader, PdfWriter
except ImportError:
    print("Instalando dependencias necesarias...")
    os.system('pip install PyPDF2')
    from PyPDF2 import PdfReader, PdfWriter

from google.colab import files
import zipfile

def split_pdf_process():
    # 1. Carga de archivo
    print("--- FASE 1: CARGA DE ARCHIVO ---")
    uploaded = files.upload()

    if not uploaded:
        print("No se subió ningún archivo. Proceso abortado.")
        return

    for filename in uploaded.keys():
        print(f"\nProcesando: {filename}")

        # 2. Preparación de directorios
        base_name = os.path.splitext(filename)[0]
        output_dir = f"split_{base_name}"
        if not os.path.exists(output_dir):
            os.makedirs(output_dir)

        # 3. Lectura y Segmentación
        try:
            reader = PdfReader(filename)
            total_pages = len(reader.pages)
            print(f"Total de hojas detectadas: {total_pages}")

            generated_files = []

            for i in range(total_pages):
                writer = PdfWriter()
                writer.add_page(reader.pages[i])

                output_filename = f"{base_name}_hoja_{i+1}.pdf"
                output_path = os.path.join(output_dir, output_filename)

                with open(output_path, "wb") as f:
                    writer.write(f)

                generated_files.append(output_path)
                print(f" -> Hoja {i+1} extraída.")

            # 4. Empaquetado (ZIP) para descarga masiva
            zip_name = f"{base_name}_hojas_separadas.zip"
            with zipfile.ZipFile(zip_name, 'w') as zipf:
                for file in generated_files:
                    zipf.write(file, os.path.basename(file))

            print(f"\n--- PROCESO COMPLETADO ---")
            print(f"Se han generado {total_pages} archivos.")
            print(f"Descargando el paquete ZIP: {zip_name}")

            # 5. Descarga automática
            files.download(zip_name)

        except Exception as e:
            print(f"Error crítico procesando el PDF: {e}")

# Ejecución del script
if __name__ == "__main__":
    split_pdf_process()
```
