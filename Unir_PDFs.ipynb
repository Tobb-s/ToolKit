```
# 1. Instalación de la librería necesaria
!pip install PyPDF2 -q

import os
from PyPDF2 import PdfWriter
from google.colab import files

def unir_pdfs():
    # 2. Subir archivos desde tu computadora
    print("--- Paso 1: Seleccioná los PDFs que querés unir ---")
    subidos = files.upload()
    nombres_archivos = list(subidos.keys())

    if len(nombres_archivos) < 2:
        return print("Error: Necesitás subir al menos dos archivos para unirlos.")

    # Ordenar alfabéticamente por defecto (o podés confiar en el orden de selección)
    nombres_archivos.sort()

    # 3. Lógica de unión
    escritor = PdfWriter()

    print(f"\nUniendo {len(nombres_archivos)} archivos...")
    for pdf in nombres_archivos:
        escritor.append(pdf)

    # 4. Definir nombre de salida
    nombre_salida = input("\n¿Qué nombre querés para el archivo final? (ej: reporte_final.pdf): ")
    if not nombre_salida.endswith('.pdf'):
        nombre_salida += '.pdf'

    # 5. Guardar y Descargar
    with open(nombre_salida, "wb") as f_salida:
        escritor.write(f_salida)

    print(f"\nListo. El archivo '{nombre_salida}' se descargará ahora.")
    files.download(nombre_salida)

# Ejecutar la función
unir_pdfs()
```
